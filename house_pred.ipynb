```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('./data/train.csv') ## 1460 rows x 81 columns
test = pd.read_csv('./data/test.csv') ## 1459 rows x 80 columns
train.head(15)
```

```python
test.head(15)
```

```python
import matplotlib.pyplot as plt

## Checking the Linearity of the Overall Quality-Sale Price and Overall Condition-Sale Price

overall_quality = train['OverallQual']
overall_condition = train['OverallCond']
sale_prices = train['SalePrice']

fig, ax = plt.subplots(figsize=(6, 5))
ax.scatter(overall_quality, sale_prices, color='blue')
ax.set_title("Overall Quality vs Sale Price")
ax.set_xlabel("Overall Quality (1-10)")
ax.set_ylabel("Sale Price")
plt.show()
```

```python
## Fitting a Mathematical Model (Polynomical Function) to the Overall Quality vs Sale Price Scatter Plot

from scipy.optimize import curve_fit

def polynomical_func(x, a, b, c, d, e):
    return a * (x**4) + b * (x**3) + c * (x**2) + d * x + e

popt, pcov = curve_fit(polynomical_func, overall_quality, sale_prices)

x_quality = np.linspace(min(overall_quality), max(overall_quality), 100)
y_quality = polynomical_func(x_quality, *popt)
```

```python
fig, ax = plt.subplots(figsize=(6, 5))

# Plot Overall Quality vs Sale Price
ax.scatter(overall_quality, sale_prices, color='blue', label="Data")
ax.plot(x_quality, y_quality, color='black', linewidth=2, label="Fitted Curve")
ax.set_title("Overall Quality vs Sale Price (Quartic Model)")
ax.set_xlabel("Overall Quality (1-10)")
ax.set_ylabel("Sale Price")
ax.legend()
```

```python
train["TotalHouseArea"] = train["1stFlrSF"] + train["2ndFlrSF"] + train["TotalBsmtSF"]
test["TotalHouseArea"] = test["1stFlrSF"] + test["2ndFlrSF"] + test["TotalBsmtSF"]

fig, ax = plt.subplots(figsize=(12, 5))
ax.scatter(train["TotalHouseArea"], train["SalePrice"], color='green')
ax.set_title("Total House Area vs Sale Price")
ax.set_xlabel("Total House Area")
ax.set_ylabel("Sale Price")
plt.show()
```

```python
## Fitting a Mathematical Model (Polynomical Function) to the Total House Area vs Sale Price Scatter Plot
total_area = train.loc[train["TotalHouseArea"] < 7500, ['TotalHouseArea']].values.flatten()
sale_price_A = train.loc[train["TotalHouseArea"] < 7500, ['SalePrice']].values.flatten()
```

```python
def polynomical_func2(x, a, b, c):
    return a * (x**2) + b * x + c

popt, pcov = curve_fit(polynomical_func2, total_area, sale_price_A)

x_fit = np.linspace(min(total_area), max(total_area), 100)
y_fit = polynomical_func2(x_fit, *popt)

print(popt)
```
