```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('./data/train.csv') ## 1460 rows x 81 columns
test = pd.read_csv('./data/test.csv') ## 1459 rows x 80 columns
train.head(15)
```

```python
test.head(15)
```

```python
import matplotlib.pyplot as plt

## Checking the Linearity of the Overall Quality-Sale Price and Overall Condition-Sale Price

overall_quality = train['OverallQual']
overall_condition = train['OverallCond']
sale_prices = train['SalePrice']

fig, ax = plt.subplots(figsize=(6, 5))
ax.scatter(overall_quality, sale_prices, color='blue')
ax.set_title("Overall Quality vs Sale Price")
ax.set_xlabel("Overall Quality (1-10)")
ax.set_ylabel("Sale Price")
plt.show()
```

```python
## Fitting a Mathematical Model (Polynomical Function) to the Overall Quality vs Sale Price Scatter Plot

from scipy.optimize import curve_fit

def polynomical_func(x, a, b, c, d, e):
    return a * (x**4) + b * (x**3) + c * (x**2) + d * x + e

popt, pcov = curve_fit(polynomical_func, overall_quality, sale_prices)

x_quality = np.linspace(min(overall_quality), max(overall_quality), 100)
y_quality = polynomical_func(x_quality, *popt)
```

```python
fig, ax = plt.subplots(figsize=(6, 5))

# Plot Overall Quality vs Sale Price
ax.scatter(overall_quality, sale_prices, color='blue', label="Data")
ax.plot(x_quality, y_quality, color='black', linewidth=2, label="Fitted Curve")
ax.set_title("Overall Quality vs Sale Price (Quartic Model)")
ax.set_xlabel("Overall Quality (1-10)")
ax.set_ylabel("Sale Price")
ax.legend()
```

```python
train["TotalHouseArea"] = train["1stFlrSF"] + train["2ndFlrSF"] + train["TotalBsmtSF"]
test["TotalHouseArea"] = test["1stFlrSF"] + test["2ndFlrSF"] + test["TotalBsmtSF"]

fig, ax = plt.subplots(figsize=(12, 5))
ax.scatter(train["TotalHouseArea"], train["SalePrice"], color='green')
ax.set_title("Total House Area vs Sale Price")
ax.set_xlabel("Total House Area")
ax.set_ylabel("Sale Price")
plt.show()
```

```python
## Fitting a Mathematical Model (Polynomical Function) to the Total House Area vs Sale Price Scatter Plot
total_area = train.loc[train["TotalHouseArea"] < 7500, ['TotalHouseArea']].values.flatten()
sale_price_A = train.loc[train["TotalHouseArea"] < 7500, ['SalePrice']].values.flatten()
```

```python
def polynomical_func2(x, a, b, c):
    return a * (x**2) + b * x + c

popt, pcov = curve_fit(polynomical_func2, total_area, sale_price_A)

x_fit = np.linspace(min(total_area), max(total_area), 100)
y_fit = polynomical_func2(x_fit, *popt)

print(popt)
```

```python
fig, ax = plt.subplots(figsize=(6, 5))
ax.scatter(total_area, sale_price_A, color='green', label="Data")
ax.plot(x_fit, y_fit, color='black', linewidth=2, label="Fitted Curve")
ax.set_title("Total House Area vs Sale Price")
ax.set_xlabel("Total House Area")
ax.set_ylabel("Sale Price")
ax.legend()
plt.show()
```

```python
from sklearn import preprocessing

categorical_features = train.select_dtypes(include=["object"]).columns

print(categorical_features) ## Both nominal and ordinal features are available.
print(len(categorical_features))
```

```python
nominal_features = [
    "MSSubClass", "MSZoning", "Street", "Alley", "LotConfig", "Neighborhood",
    "Condition1", "Condition2", "BldgType", "HouseStyle", "RoofStyle", "RoofMatl",
    "Exterior1st", "Exterior2nd", "Foundation", "Heating", "CentralAir",
    "Electrical", "GarageType", "MiscFeature", "SaleType", "SaleCondition"
]

ordinal_features = {
    "LotShape": ["Reg", "IR1", "IR2", "IR3"],
    "LandContour": ["Lvl", "Bnk", "HLS", "Low"],
    "LandSlope": ["Gtl", "Mod", "Sev"],
    "OverallQual": list(range(1, 11)),
    "OverallCond": list(range(1, 11)),
    "ExterQual": ["Po", "Fa", "TA", "Gd", "Ex"],
    "ExterCond": ["Po", "Fa", "TA", "Gd", "Ex"],
    "BsmtQual": ["NA", "Po", "Fa", "TA", "Gd", "Ex"],
    "BsmtCond": ["NA", "Po", "Fa", "TA", "Gd", "Ex"],
    "BsmtExposure": ["NA", "No", "Mn", "Av", "Gd"],
    "BsmtFinType1": ["NA", "Unf", "LwQ", "Rec", "BLQ", "ALQ", "GLQ"],
    "BsmtFinType2": ["NA", "Unf", "LwQ", "Rec", "BLQ", "ALQ", "GLQ"],
    "HeatingQC": ["Po", "Fa", "TA", "Gd", "Ex"],
    "KitchenQual": ["Po", "Fa", "TA", "Gd", "Ex"],
    "FireplaceQu": ["NA", "Po", "Fa", "TA", "Gd", "Ex"],
    "GarageFinish": ["NA", "Unf", "RFn", "Fin"],
    "GarageQual": ["NA", "Po", "Fa", "TA", "Gd", "Ex"],
    "GarageCond": ["NA", "Po", "Fa", "TA", "Gd", "Ex"],
    "PavedDrive": ["N", "P", "Y"],
    "PoolQC": ["NA", "Fa", "TA", "Gd", "Ex"],
    "Fence": ["NA", "MnWw", "GdWo", "MnPrv", "GdPrv"],
    "Functional": ["Sal", "Sev", "Maj2", "Maj1", "Mod", "Min2", "Min1", "Typ"]
}
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

one_hot_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
encoded_nominal_data = one_hot_encoder.fit_transform(train[nominal_features])

encoded_nominal_df = pd.DataFrame(encoded_nominal_data, columns=one_hot_encoder.get_feature_names_out(nominal_features))
encoded_nominal_df.head(10)
```

```python
for feature in ordinal_features:
    train[feature] = train[feature].fillna(ordinal_features[feature][0])

ordinal_encoder = OrdinalEncoder(categories=[ordinal_features[feature] for feature in ordinal_features], dtype=int)
encoded_ordinal_data = ordinal_encoder.fit_transform(train[list(ordinal_features.keys())])
encoded_ordinal_df = pd.DataFrame(encoded_ordinal_data, columns=ordinal_features.keys())
print(encoded_ordinal_df.head(10))
```

```python
print(encoded_nominal_df.head(10))
```

```python
# Drop original categorical columns and concatenate encoded data
train_encoded_df = train.drop(columns=nominal_features + list(ordinal_features.keys())).reset_index(drop=True)
train_encoded_df = pd.concat([train_encoded_df, encoded_nominal_df, encoded_ordinal_df], axis=1)
print(train_encoded_df.head(10))
```
