```python
train = pd.read_csv('./data/train.csv') ## 1460 rows x 81 columns
test = pd.read_csv('./data/test.csv') ## 1459 rows x 80 columns
train.head(15)
```

```python
import matplotlib.pyplot as plt

## Checking the Linearity of the Overall Quality-Sale Price and Overall Condition-Sale Price

overall_quality = train['OverallQual']
overall_condition = train['OverallCond']
sale_prices = train['SalePrice']

fig, ax = plt.subplots(figsize=(6, 5))
ax.scatter(overall_quality, sale_prices, color='blue')
ax.set_title("Overall Quality vs Sale Price")
ax.set_xlabel("Overall Quality (1-10)")
ax.set_ylabel("Sale Price")
plt.show()
```

```python
## Fitting a Mathematical Model (Polynomical Function) to the Overall Quality vs Sale Price Scatter Plot

from scipy.optimize import curve_fit

def polynomical_func(x, a, b, c, d, e):
    return a * (x**4) + b * (x**3) + c * (x**2) + d * x + e

popt, pcov = curve_fit(polynomical_func, overall_quality, sale_prices)

x_quality = np.linspace(min(overall_quality), max(overall_quality), 100)
y_quality = polynomical_func(x_quality, *popt)
```

```python
fig, ax = plt.subplots(figsize=(6, 5))

# Plot Overall Quality vs Sale Price
ax.scatter(overall_quality, sale_prices, color='blue', label="Data")
ax.plot(x_quality, y_quality, color='black', linewidth=2, label="Fitted Curve")
ax.set_title("Overall Quality vs Sale Price (Quartic Model)")
ax.set_xlabel("Overall Quality (1-10)")
ax.set_ylabel("Sale Price")
ax.legend()
```

```python
train["TotalHouseArea"] = train["1stFlrSF"] + train["2ndFlrSF"] + train["TotalBsmtSF"]
test["TotalHouseArea"] = test["1stFlrSF"] + test["2ndFlrSF"] + test["TotalBsmtSF"]

fig, ax = plt.subplots(figsize=(12, 5))
ax.scatter(train["TotalHouseArea"], train["SalePrice"], color='green')
ax.set_title("Total House Area vs Sale Price")
ax.set_xlabel("Total House Area")
ax.set_ylabel("Sale Price")
plt.show()
```

```python
## Fitting a Mathematical Model (Polynomical Function) to the Total House Area vs Sale Price Scatter Plot
total_area = train.loc[train["TotalHouseArea"] < 7500, ['TotalHouseArea']].values.flatten()
sale_price_A = train.loc[train["TotalHouseArea"] < 7500, ['SalePrice']].values.flatten()
```

```python
def polynomical_func2(x, a, b, c):
    return a * (x**2) + b * x + c

popt, pcov = curve_fit(polynomical_func2, total_area, sale_price_A)

x_fit = np.linspace(min(total_area), max(total_area), 100)
y_fit = polynomical_func2(x_fit, *popt)

print(popt)
```

```python
fig, ax = plt.subplots(figsize=(6, 5))
ax.scatter(total_area, sale_price_A, color='green', label="Data")
ax.plot(x_fit, y_fit, color='black', linewidth=2, label="Fitted Curve")
ax.set_title("Total House Area vs Sale Price")
ax.set_xlabel("Total House Area")
ax.set_ylabel("Sale Price")
ax.legend()
plt.show()
```

```python
## Splitting Nominal and Ordinal Features

nominal_features = [
    "MSSubClass", "MSZoning", "Street", "Alley", "LotConfig", "Neighborhood",
    "Condition1", "Condition2", "BldgType", "HouseStyle", "RoofStyle", "RoofMatl",
    "Exterior1st", "Exterior2nd", "Foundation", "Heating", "CentralAir",
    "Electrical", "GarageType", "MiscFeature", "SaleType", "SaleCondition", "Utilities", "MasVnrType"
]

ordinal_features = {
    "LotShape": ["Reg", "IR1", "IR2", "IR3"],
    "LandContour": ["Lvl", "Bnk", "HLS", "Low"],
    "LandSlope": ["Gtl", "Mod", "Sev"],
    "OverallQual": list(range(1, 11)),
    "OverallCond": list(range(1, 11)),
    "ExterQual": ["Po", "Fa", "TA", "Gd", "Ex"],
    "ExterCond": ["Po", "Fa", "TA", "Gd", "Ex"],
    "BsmtQual": ["NA", "Po", "Fa", "TA", "Gd", "Ex"],
    "BsmtCond": ["NA", "Po", "Fa", "TA", "Gd", "Ex"],
    "BsmtExposure": ["NA", "No", "Mn", "Av", "Gd"],
    "BsmtFinType1": ["NA", "Unf", "LwQ", "Rec", "BLQ", "ALQ", "GLQ"],
    "BsmtFinType2": ["NA", "Unf", "LwQ", "Rec", "BLQ", "ALQ", "GLQ"],
    "HeatingQC": ["Po", "Fa", "TA", "Gd", "Ex"],
    "KitchenQual": ["Po", "Fa", "TA", "Gd", "Ex"],
    "FireplaceQu": ["NA", "Po", "Fa", "TA", "Gd", "Ex"],
    "GarageFinish": ["NA", "Unf", "RFn", "Fin"],
    "GarageQual": ["NA", "Po", "Fa", "TA", "Gd", "Ex"],
    "GarageCond": ["NA", "Po", "Fa", "TA", "Gd", "Ex"],
    "PavedDrive": ["N", "P", "Y"],
    "PoolQC": ["NA", "Fa", "TA", "Gd", "Ex"],
    "Fence": ["NA", "MnWw", "GdWo", "MnPrv", "GdPrv"],
    "Functional": ["Sal", "Sev", "Maj2", "Maj1", "Mod", "Min2", "Min1", "Typ"]
}
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

## One-Hot Encoding for Nominal Features
one_hot_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False) 
encoded_nominal_data = one_hot_encoder.fit_transform(train[nominal_features]) ## Creates a binary matrix for each nominal feature and its classes.
encoded_nominal_data_test = one_hot_encoder.transform(test[nominal_features])

print(one_hot_encoder.get_feature_names_out(nominal_features))
```

```python
print(encoded_nominal_data.shape)
```

```python
encoded_nominal_df = pd.DataFrame(encoded_nominal_data, columns=one_hot_encoder.get_feature_names_out(nominal_features))
encoded_nominal_df_test = pd.DataFrame(encoded_nominal_data_test, columns=one_hot_encoder.get_feature_names_out(nominal_features))
encoded_nominal_df.head(10)
print(encoded_nominal_df.shape)
```

```python
## Ordinal Encoding for Ordinal Features

for feature in ordinal_features:
    train[feature] = train[feature].fillna(ordinal_features[feature][0])
    test[feature] = test[feature].fillna(ordinal_features[feature][0])


ordinal_encoder = OrdinalEncoder(categories=[ordinal_features[feature] for feature in ordinal_features], dtype=int) ## Passing the categories of each ordinal feature to the OrdinalEncoder.
encoded_ordinal_data = ordinal_encoder.fit_transform(train[list(ordinal_features.keys())]) ## Creating a matrix for each ordinal feature and its classes.
encoded_ordinal_data_test = ordinal_encoder.transform(test[list(ordinal_features.keys())])

encoded_ordinal_df = pd.DataFrame(encoded_ordinal_data, columns=ordinal_features.keys())
encoded_ordinal_df_test = pd.DataFrame(encoded_ordinal_data_test, columns=ordinal_features.keys())

print(encoded_ordinal_df.loc[0:40, ["Functional"]])
```

```python
print(encoded_ordinal_df.shape)
print(encoded_ordinal_df.head(5))
```

```python
# Drop original categorical columns and concatenate encoded Categorical Features with Numerical Features
train_encoded_df = train.drop(columns=nominal_features + list(ordinal_features.keys())).reset_index(drop=True)
test_encoded_df = test.drop(columns=nominal_features + list(ordinal_features.keys())).reset_index(drop=True)
print(train_encoded_df.shape)
print(train_encoded_df.head(5))
```

```python
train_encoded_df = pd.concat([train_encoded_df, encoded_nominal_df, encoded_ordinal_df], axis=1)
test_encoded_df = pd.concat([test_encoded_df, encoded_nominal_df_test, encoded_ordinal_df_test], axis=1)

train_encoded_df = train_encoded_df.drop(columns=["TotalHouseArea"])
test_encoded_df = test_encoded_df.drop(columns=["TotalHouseArea"])

print(train_encoded_df.shape)
print(train_encoded_df.head(5))
```

```python
train_encoded_df.to_csv('./data/train_encoded.csv', index=False)
test_encoded_df.to_csv('./data/test_encoded.csv', index=False)
```

```python
df = pd.read_csv('./data/train_encoded.csv')
df_test = pd.read_csv('./data/test_encoded.csv')

df = df.drop(columns=["Id"])
df_test = df_test.drop(columns=["Id"])
```

```python
## Treating the Missing Values in the Dataset

df["LotFrontage"] = pd.to_numeric(df["LotFrontage"], errors="coerce")
df["MasVnrArea"] = pd.to_numeric(df["MasVnrArea"], errors="coerce")
df["GarageYrBlt"] = pd.to_numeric(df["GarageYrBlt"], errors="coerce")

df_test["LotFrontage"] = pd.to_numeric(df_test["LotFrontage"], errors="coerce")
df_test["MasVnrArea"] = pd.to_numeric(df_test["MasVnrArea"], errors="coerce")
df_test["GarageYrBlt"] = pd.to_numeric(df_test["GarageYrBlt"], errors="coerce")

# Apply missing value imputation safely
df.fillna({
    "LotFrontage": df["LotFrontage"].median() if not df["LotFrontage"].isnull().all() else 0,
    "MasVnrArea": df["MasVnrArea"].median() if not df["MasVnrArea"].isnull().all() else 0,
    "GarageYrBlt": df["YearBuilt"].median() if "YearBuilt" in df.columns and not df["YearBuilt"].isnull().all() else 0
}, inplace=True)

df_test.fillna({
    "LotFrontage": df_test["LotFrontage"].median() if not df_test["LotFrontage"].isnull().all() else 0,
    "MasVnrArea": df_test["MasVnrArea"].median() if not df_test["MasVnrArea"].isnull().all() else 0,
    "GarageYrBlt": df_test["YearBuilt"].median() if "YearBuilt" in df_test.columns and not df_test["YearBuilt"].isnull().all() else 0
}, inplace=True)

numeric_cols = df_test.select_dtypes(include=['number']).columns

for col in numeric_cols:
    if df_test[col].isnull().sum() > 0:
        df_test[col].fillna(df_test[col].median(), inplace=True)

# Verify missing values are handled
print(df.isnull().sum().sum())  # Should print 0
print(df_test.isnull().sum().sum())  # Should print 0
```
