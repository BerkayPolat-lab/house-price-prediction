```python
import matplotlib.pyplot as plt

## Checking the Linearity of the Overall Quality-Sale Price and Overall Condition-Sale Price

overall_quality = train['OverallQual']
overall_condition = train['OverallCond']
sale_price = train['SalePrice']

fig, ax = plt.subplots(figsize=(6, 5))
ax.scatter(overall_quality, sale_price, color='blue')
ax.set_title("Overall Quality vs Sale Price")
ax.set_xlabel("Overall Quality (1-10)")
ax.set_ylabel("Sale Price")
plt.show()
```

```python
## Fitting a Mathematical Model (Polynomical Function) to the Overall Quality vs Sale Price Scatter Plot

from scipy.optimize import curve_fit

def polynomical_func(x, a, b, c, d, e):
    return a * (x**4) + b * (x**3) + c * (x**2) + d * x + e

popt, pcov = curve_fit(polynomical_func, overall_quality, sale_price)

x_quality = np.linspace(min(overall_quality), max(overall_quality), 100)
y_quality = polynomical_func(x_quality, *popt)
```

```python
fig, ax = plt.subplots(figsize=(6, 5))

# Plot Overall Quality vs Sale Price
ax.scatter(overall_quality, sale_price, color='blue', label="Data")
ax.plot(x_quality, y_quality, color='black', linewidth=2, label="Fitted Curve")
ax.set_title("Overall Quality vs Sale Price")
ax.set_xlabel("Overall Quality (1-10)")
ax.set_ylabel("Sale Price")
ax.legend()
```
